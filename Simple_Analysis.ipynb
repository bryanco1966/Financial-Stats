### Returns
    Simple returns are calculated taking the Price at time t minus the price at time 0.  Then divided by time zero

\begin{equation*}
Return =   \frac{Final Price - Initial Price}{Initial Price}
\end{equation*}

### Gross Returns
    Gross returns are simple returns where you subtract 1 from both sides of the equation.

\begin{equation*}
Gross Return = \frac{Final Price}{Initial Price}
\end{equation*}

### Log Return
    The log return is the log of the Gross return.  This is convien  
    
\begin{equation*}
Log(Gross Return) =Log\frac{Final Price}{Initial Price} = Log(Final Price) - Log(Initial Price)
\end{equation*}
