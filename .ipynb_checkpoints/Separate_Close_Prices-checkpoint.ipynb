#  Basic Analysis

    I will do some basic analysis on the technical data of stocks using the close prices and the end of each day.
    to do that I will have to reorder the S and P dataframe to have the date as an index and each row be a stocks 
    closing price with the column name being the stock.

```python
import pandas as pd
```

```python
filename = '../data/sp500full.csv'
ps = pd.read_csv(filename)
```

```python
ps.info()
```

```python
dates = pd.date_range(ps['Date'].min(), ps['Date'].max())
```

```python
dates
```

```python
#create new dataframe with the indexes being these dates
close_prices = pd.DataFrame(index = dates)
```

```python
close_prices
```

```python
#to add columns
symbols = ps['Stock'].unique()

for symbol in symbols:
    df_sym = ps[ps['Stock']==symbol]
    #data parameter must be an ndarray iterable, dict or dataframe. 
    df_temp = pd.DataFrame(data = df_sym['Close'].to_numpy(),
                          index = df_sym['Date'], columns = [symbol])
    close_prices = close_prices.join(df_temp)
```

```python
#drop weekends and holidays
close_prices.dropna(axis = 0, how = 'all', inplace = True)
```

```python
#first forward fill then backfill the data
close_prices.fillna(method = 'ffill', inplace = True)
close_prices.fillna(method = 'bfill', inplace = True)
```

```python
#check to make sure no more missing data
close_prices.isna().sum().sum()
```

```python
#normalize and set mean to 100
close_prices = (close_prices/close_prices.iloc[0])*100
```

```python
close_prices.plot(legend= False, figsize = (10,10))
```

```python
high = close_prices[close_prices>5000].dropna(axis = 1, how = 'all')
high.plot()
```

```python
filename = '../data/close_prices.csv'
close_prices.to_csv(filename, index_label='Date')
```
