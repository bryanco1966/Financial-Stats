### Returns
    Simple returns are calculated taking the Price at time t minus the price at time 0.  Then divided by time zero

\begin{equation*}
Return =   \frac{Final Price - Initial Price}{Initial Price}
\end{equation*}

### Gross Returns
    Gross returns are simple returns where you subtract 1 from both sides of the equation.

\begin{equation*}
Gross Return = \frac{Final Price}{Initial Price}
\end{equation*}

### Log Return
    The log return is the log of the Gross return.  This is convenient because the log gross returns can be 
    calculated using simple subtraction rather than multiplication and division
    
\begin{equation*}
Log(Gross Return) =Log\frac{Final Price}{Initial Price} = Log(Final Price) - Log(Initial Price)
\end{equation*}

### QQPlot Code 

Also referred to as the normal probability plot. Real world returns tend to be more fat tailed as in normal plot
extremes are very rare.  If QQplot is a straight line it will indicate normal distribution fat tails will diverge from the lines at both ends.

```python
stbux = pd.read_csv(filename, index_col= 'Date', usecols=['Date','SBUX'], parse_dates = True)
```

```python
stbux['return']= stbux['SBUX'].pct_change()
```

```python
#linspace creates a list of evenly spaced numbers over an interval 
#so we create 100 evenly spaced number between minimum and maximum return
#this really just creates a x axis for our plot
x_list = np.linspace(stbux['return'].min(), stbux['return'].max(),100)
```

```python
# create a normal distribution using mean and std of starbux returns
y_list = norm.pdf(x_list, loc=stbux['return'].mean(), scale = stbux['return'].std())
```

```python
#plot the normal return and a histogram of the starbux data
plt.plot(x_list, y_list)
# density = True will normalize the values of the histogram.  Raw values will be much larger
stbux['return'].hist(bins= 100, density = True)
```

#### QQplot showing this distribution

    This plot definitely shows the fat tails of the returns as the diverge from the normal line

```python
probplot(stbux['return'].dropna(), dist = 'norm', fit = True, plot = plt);
```

### Look at t-ditribution 

    The tricky thing about the t-distribution is that it has three parameters to calculate.  The normal function only needs the mean and the standard deviation to calculate it.  The t distribution cannot be estimated in its closed form and must be estimated using maximum likelihood analysis.  
    
\begin{equation*}
t =\frac{X-u}{s/sqrt(n)} 
\end{equation*}  
    
    
So in this case we are finding out how much the sample mean diverges from the popultion mean and we have three variables. The sample mean, sample standard deviation, and the degrees of freedom.    

```python
params = t.fit(stbux['return'].dropna())
```

```python
stbux['return'].mean()
```

```python
len(stbux['return'])
```

```python
params
```

```python
#careful I am reassigning some variables here df used to be a dataframe
df, loc, scale = params
```

```python
y_list = t.pdf(x_list, df, loc, scale)
```

```python
plt.plot(x_list, y_list)
plt.title('Compare Starbux Returns to T-Distribution')
plt.xlabel('Return')
plt.ylabel('Number of Occurances')
stbux['return'].hist(bins= 100, density = True);
```

```python
#build custom class distribution for t-distribution beacuse we need degrees for freedom.
class MyT:
    
    def __init__(self, df):
        self.df = df
        
    def fit(self,x):
        return t.fit(x)
    
    #ppf defaults to mean 0 and standard deviation 1
    def ppf(self, x, loc = 0, scale=1):
        return t.ppf(x, self.df, loc, scale)
        
```

```python
sm.qqplot(stbux['return'].dropna(), dist = MyT(df), line = 's')
plt.title('QQplot for t-distribution Starbucks Returns');
```

### Skewness and Kurtosis

The third and fourth moment of a random distribution


\begin{align}
Skewness & = E[(\frac{\bar{X} -\mu)}{\sigma}^3]\\
Kurtosis & = E[(\frac{\bar{X} -\mu)}{\sigma}^4]\\
\end{align}

Skewness is the symmetry of a distribution left or negative skew has a tail on the left while right or positive skewness has a tail on the right.  In financial analysis we want a positive skew with a heavy tail in the positive direction

Kurtosis measures how heavy the tails are.  A normal distribution has a kurtosis of 3 and therefore it is often discusses as the amount of excess kurtosis.  Are the tails heavier or lighter than a standard distribution.  Generally in finance a high kurtosis is equated with a high risk.  Making possible to both win and lose big.  Therefore, it is often avoided, but might have value in certain instances.

Just a side note when calculating skewness or kurtosis from a sample mean dividing by the degrees of freedom give you the unbiased estimate while the Maximum Likelihood estimate below would use us the sample size.

\begin{align}
Sample Skewness & = 1/N\sum_{i=1}^{N}\frac{(x_i -\bar{X})}{s}^3\\
Sample Kurtosis & = 1/N\sum_{i=1}^{N}\frac{(x_i -\bar{X})}{s}^4\\
\end{align}

```python
stbux['return'].skew()
```

```python
#this is the excess kurtosis showing it has heavy tails
stbux['return'].kurtosis()
```

## Confidence Intervals

Graphing techniques to show confidence intervals of different distributions.  More data you collect the smaller the confidence intervals. We are attempting to find the true return and if our sample is positive it does not mean the true mean is positive.  A confidence interval attempts to determine the chance the return is positive.  

Of course since this is a time series other things are happening that could affect the return over time so the sample return might be a little problematic to begin with.

\begin{align}
Confidence Interval& = -1.96(\frac{\sigma}{\sqrt{n}}) > \mu > 1.96(\frac{\sigma}{\sqrt{n}})\\
\end{align}

We do not know sigma so we use sample standard deviation which is t distributed and approaches normal as n goes to infinity.

```python
values = stbux['return'].dropna().to_numpy()
```

```python
values_mean = values.mean()
values_sd = values.std(ddof = 1)
```

```python
low_estimate = values_mean -1.96 *(values_sd/np.sqrt(len(values)))
high_estimate =values_mean + 1.96* (values_sd/np.sqrt(len(values)))
```

```python

plt.axvline(values_mean, label= "Mean", color = 'red')
plt.axvline(low_estimate, label= "CI", color = 'green')
plt.axvline(high_estimate, label= "CI", color = 'green')
plt.xlim(1e-5, 3e-3)
plt.legend();
```

```python
low_estimate, high_estimate, values_mean, values_sd
```

### Test for normality

We will import the Jacque-Berra and normality test to determine if the distribution of retruns are normal.  We assume that they will not be since they have heavy tails and will test to see if the distribution is t-distributed as well. 

```python
from scipy.stats import jarque_bera, normaltest, kstest, ttest_1samp
```

```python
# values is the starbux returns minus the missing values we created above.
# a significant result means we can reject the null hypothesis that the distribution
#is normal.
normaltest(values)
```

```python
jarque_bera(values)
```

```python
df, loc, scale = t.fit(values)

def cdf(x):
    return t.cdf(x,df,loc,scale)

#significant result would suggest different than a t-distribution
kstest(values,cdf)
```

```python
#use ttest to determine if different form zero
ttest_1samp(values,0)
```
